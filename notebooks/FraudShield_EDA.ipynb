# 🛡️ FraudShield AI — Exploratory Data Analysis
This notebook mirrors the analysis performed in the main platform.

```python
import sys, os
sys.path.insert(0, '..')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from utils.data_generator import generate_fraud_dataset

sns.set_style('darkgrid')
plt.rcParams.update({'figure.facecolor': '#0f172a', 'axes.facecolor': '#1e293b',
                     'text.color': '#e2e8f0', 'axes.labelcolor': '#e2e8f0',
                     'xtick.color': '#94a3b8', 'ytick.color': '#94a3b8'})
```

```python
df = generate_fraud_dataset(10000)
print(f'Shape: {df.shape}')
print(f'Fraud Rate: {df["Fraud_Label"].mean():.2%}')
df.head()
```

```python
df.info()
df.describe()
```

```python
# Fraud distribution
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
df['Fraud_Label'].value_counts().plot.pie(ax=axes[0], autopct='%1.1f%%',
    colors=['#22c55e', '#ef4444'], labels=['Legitimate', 'Fraudulent'])
axes[0].set_title('Class Distribution')
sns.histplot(data=df, x='Transaction_Amount', hue='Fraud_Label',
             palette={0:'#22c55e', 1:'#ef4444'}, bins=50, ax=axes[1])
axes[1].set_title('Amount Distribution by Class')
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
num_cols = ['Transaction_Amount', 'Account_Balance', 'Daily_Transaction_Count', 'Card_Age',
            'Previous_Fraudulent_Activity', 'Fraud_Label']
plt.figure(figsize=(8, 5))
sns.heatmap(df[num_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()
```
